```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("UCI_Credit_Card.csv")
df
```

```python
for i in range(25):
    print("The number of null values in ",i,"are:")
    nullcheck=df.iloc[:,i].isnull().sum()
    print(nullcheck)
```

```python
Y = df['default.payment.next.month']
Y
```

```python
X=df.drop(['ID','default.payment.next.month'],axis= 'columns')
X
```

```python
def normalize(X):
    X-np.mean(X) / np.std(X)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X.iloc[:,12:24]=scaler.fit_transform(X.iloc[:,12:24])
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=7)
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
Y_predicted = lr.predict(X_test)
print('The first five predicted values are:')
print(Y_predicted[0:5])
print('The first five test values are:')
print(np.array(Y_test[0:5]))
print('The accuracy rate of LogisticRegression is:')
print(lr.score(X_test,Y_test))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Scale the training data

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, Y_train)
```

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(Y_predicted,Y_test)
sns.heatmap(matrix,annot=True,cmap='Greens',fmt='0.1f')
plt.xlabel('Predicted Values')
plt.ylabel('Test Values')
plt.title('Confusion matrix for LogisticRegression')
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,Y_train)
Y_predicted=nb.predict(X_test)
print('The first five predicted values are:')
print(Y_predicted[0:5])
print('The first five test values are:')
print(np.array(Y_test[0:5]))
print('The accuracy rate of NaiveBayesClassifier is:')
print(lr.score(X_test,Y_test))
```
